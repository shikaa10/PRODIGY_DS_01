```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_1339682.csv'
data = pd.read_csv(file_path, skiprows=4)

# Filter the data for India
india_data = data[data['Country Name'] == 'India'].dropna(axis=1, how='all')

# Extract years and population values
years = india_data.columns[4:-1]
population = india_data.iloc[0, 4:-1].values

# Convert years to integers for plotting
years = years.astype(int)

# Plot the bar chart
plt.figure(figsize=(14, 8))
plt.bar(years, population, color='skyblue')

# Add titles and labels
plt.title('Population of India (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)

# Display the chart
plt.tight_layout()
plt.show()
```
